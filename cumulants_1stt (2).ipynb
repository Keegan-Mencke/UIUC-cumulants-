```python
# bloody ai way

import numpy as np

def compute_Qn(phi, n):
    """
    Compute Q_n = sum(exp(i * n * phi_i))
    """
    return np.sum(np.exp(1j * n * phi))

def compute_four_particle(phi, n):
    """
    Compute <4> based on the given formula.
    """
    M = len(phi)
    if M<=3:
        return 0
    Qn = compute_Qn(phi, n)
    Q2n = compute_Qn(phi, 2 * n)
    
    # Compute magnitudes
    abs_Qn_sq = np.abs(Qn)**2
    abs_Qn_4 = abs_Qn_sq**2
    abs_Q2n_sq = np.abs(Q2n)**2
    
    # Real part term
    real_term = np.real(Q2n * np.conj(Qn) * np.conj(Qn))
    
    # Denominator
    denom = M * (M - 1) * (M - 2) * (M - 3)
    
    # First big fraction
    term1 = (abs_Qn_4 + abs_Q2n_sq - 2 * real_term) / denom
    
    # Second big fraction
    term2 = 2 * (2*(M - 2) * abs_Qn_sq - M * (M - 3)) / denom
    
    return term1 - term2

# Example usage:
'''phi = np.random.uniform(0, 2*np.pi, 100)  # 100 random angles
n = 2  # harmonic number
result = compute_four_particle(phi, n)
print(f"<4> = {result}")'''
n=2
import time
cumulant2=np.zeros(50, dtype=complex )
#cumulant2=np.zeros(100 ) 
sum_weight = np.sum(weights ) 
start_time = time.time() 
for i in range(50):
    store1 = np.array(vtrackphi_test[i][0])  
    cumulant2[i] = compute_four_particle(store1, n) 
cumulant2[i]*weights/sum_weight
end_time = time.time() 
print(cumulant2)
print(f"Total runtime: {end_time - start_time} seconds") 
```
